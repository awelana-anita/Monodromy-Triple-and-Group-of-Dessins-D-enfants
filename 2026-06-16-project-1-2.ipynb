```sage
#Lost Code for Sigma-0,Sigma_1 and Sigma_infinity

def sigma_0_computeF(Graph):
    labels = []
    result = []
    black_vertices = Graph.left
    if not Graph.has_loops() and not Graph.has_multiple_edges():
        Graph.is_planar(set_embedding = True)
        embed = Graph.get_embedding()
    for b in black_vertices:
        if Graph.has_multiple_edges():
            for black_e in (Graph.edges_incident(b)):
                result.append(black_e[2])
        elif Graph.has_loops():
            for black_e in (Graph.edges_incident(b)):
                if black_e[0] == black_e[1]:
                    result.append(black_e[2])
            for black_v in reversed(embed[b]):
                if black_v == b:
                       continue
                label = Graph.edge_label(b,black_v)
                result.append(label)
        else:
            for black_v in reversed(embed[b]):
                label = Graph.edge_label(b,black_v)
                result.append(label)
        labels.append(tuple(result))
        result.clear()
                               
    return labels

def sigma_1_computeF(Graph):
    labels = []
    result = []
    red_vertices = Graph.right
    if not Graph.has_loops() and not Graph.has_multiple_edges():
        Graph.is_planar(set_embedding = True)
        embed = Graph.get_embedding()
    for r in red_vertices:
        if Graph.has_multiple_edges():
            for red_e in (Graph.edges_incident(r)):
                result.append(red_e[2])
        elif Graph.has_loops():
            for red_e in (Graph.edges_incident(r)):
                if red_e[0] == red_e[1]:
                    result.append(int(red_e[2]))
            for red_v in reversed(embed[r]):
                if red_v == r:
                    continue
                label = Graph.edge_label(r,red_v)
                result.append(label)
        else:
            for red_v in reversed(embed[r]):
                label = Graph.edge_label(r,red_v)
                result.append(label)
        labels.append(tuple(result))
        result.clear()
                               
    return labels












    


def sigma_infinity_computeF(Graph) :  
    if Graph.has_loops() or Graph.has_multiple_edges():
        sigma_0 = Permutation(convert_to_cycle(sigma_0_computeF(Graph)))
        sigma_1 = Permutation(convert_to_cycle(sigma_1_computeF(Graph)))
        #print(sigma_0) 
        #print(sigma_1)
        labels = (sigma_0 ^ (-1))  *( sigma_1  ^ (-1) ) 
        labels = labels.cycle_tuples()
        
    else:    
        graph_faces = Graph.faces()
        red_vertices = Graph.right
        black_vertices = Graph.left
        #print(black_vertices)
        labels = []
        result = []
        #loop through each individual face in the faces list
        for face in graph_faces:
            #print("faces:",face)
            #loop through each edge in a face
            for edge in face:
                #print("edge",edge)
                #if the first vertex of an edge is black skip
                if edge[0] in black_vertices:
                     continue
                    #print("first edge: " ,edge[0]) 
                #if the first vertex of an edge is not black, find the labeling of the edge
                else:
                    #edge_label returns the label of an edge
                    #labelled = edge[2]
                    labelled = Graph.edge_label(edge[0] ,edge[1])
                    #print("LABEL:" , labelled)
                #append the label to a result list
                    result.append(labelled)
            labels.append(tuple(result))
            
            result.clear() 
   
        
       
    return labels




def convert_to_cycle(cycles):
    cycles_int = []   
    for cycle in cycles:
        
        #if the cycel is fixed
        if not isinstance(cycle,(list,tuple)):
            cycle = (cycle,)
        new_cycle = []
        for label in cycle:    
        #if the label is in the form of a list
            if isinstance(label,list):
                label = label[0]
               
            new_cycle.append(int(str(label).strip()))
        cycles_int.append(tuple(new_cycle))
    return cycles_int
        

    
    
def check_sigma_infty(Graph):    
    labels = []
    sigma_0 = Permutation(convert_to_cycle(sigma_0_computeF(Graph)))
    sigma_1 = Permutation(convert_to_cycle(sigma_1_computeF(Graph)))
    #print(sigma_0) 
    #print(sigma_1)
    labels = (sigma_0^ (-1))  *( sigma_1  ^ (-1) )  
    return labels.cycle_tuples()


def id_check(Graph):
    labels = []
    sigma_0 = Permutation(convert_to_cycle(sigma_0_computeF(Graph)))
    print("order1:" , sigma_0.order())
    sigma_1 = Permutation(convert_to_cycle(sigma_1_computeF(Graph)))
    print("order1:" , sigma_1.order())
    sigma_infinity = Permutation(convert_to_cycle(sigma_infinity_computeF(Graph)))
    print("order1:" , sigma_infinity.order())
    #print(sigma_0) 
    #print(sigma_1)
    labels =  ( sigma_infinity )  * ( sigma_1 ) * (sigma_0) 
    return labels.cycle_tuples()
```
